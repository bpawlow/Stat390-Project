# Data Cleaning

```python
## Import necessary packages

import sklearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
## Read v4 csv dataset

data = pd.read_csv("train.csv")

# Plot box and violin plots
sns.boxenplot(data=data)
sns.violinplot(data=data)

# Add axis labels and title
plt.xlabel('Dimension')
plt.ylabel('Value')
plt.title('Distribution of Multidimensional Data')

# Show the plot
plt.show()
```

```python
#Flatten dataset to only contain unique observations 

#list of column names excluding category name 

columns = list(data.columns)
columns.remove('category_name')

# Delete duplicate rows based on specific columns 
freq_data = data.drop_duplicates(subset=columns, ignore_index = True)

display(freq_data)
```

```python
# Discard programs where the minimum age is over 25

display(freq_data[freq_data['min_age'] > 25])

under_25_data = freq_data[freq_data['min_age'] <= 25]
```

```python
# univariate table for diversity indices

div_ind_data = under_25_data.drop_duplicates(subset=['geographic_cluster_name'])

##testing cluster names corresponding to same diversity index values

test = under_25_data.drop_duplicates(subset=['gini_diversity_index'])

test_2 = under_25_data.drop_duplicates(subset=['simpson_diversity_index'])

if div_ind_data.shape[0] == test.shape[0] == test_2.shape[0]:
    print('cluster names correspond to same diversity index values')

# calculate summary statistics for columns gini and simpson indices
summary_stats = pd.concat([div_ind_data['gini_diversity_index'].describe(),
                           div_ind_data['simpson_diversity_index'].describe()], axis=1)

# create a new dataframe to store the summary statistics
summary_table = pd.DataFrame(summary_stats)

# transpose the table to make it more readable
summary_table = summary_table.transpose()

# remove the count column (as it is redundant)
summary_table = summary_table.drop('count', axis=1)

# display the clean summary statistics table
display(summary_table)
```

##Gini 

0 to 0.2: Indicates low diversity.
0.2 to 0.4: Indicates moderate diversity.
0.4 to 0.6: Indicates high diversity.
0.6 to 0.8: Indicates very high diversity.
0.8 to 1: Indicates extreme inequality.

##Simpson 

0 to 0.1: Indicates very low diversity, with one or a few dominant species and very few rare species.
0.1 to 0.3: Indicates low diversity, with a few dominant species and some rare species.
0.3 to 0.5: Indicates moderate diversity, with several dominant species and some rare species.
0.5 to 0.7: Indicates high diversity, with many species having similar abundances.
0.7 to 0.9: Indicates very high diversity, with many rare species and no or few dominant species.
0.9 to 1: Indicates maximum diversity, where each individual belongs to a different species.

```python
##check if there are any missing values

if under_25_data['capacity'].isnull().any():
    print("There are missing values in column 'capacity'")
else:
    print("There are no missing values in column 'capacity'")

# IQR test code

# calculate IQR and bounds
q1 = under_25_data['capacity'].describe()['25%']
q3 = under_25_data['capacity'].describe()['75%']
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# filter out outliers
capac_data = freq_data.loc[(freq_data['capacity'] >= lower_bound) 
            & (freq_data['capacity'] <= upper_bound)]
```

```python
# Summary statistics of capacity levels for programs, capacity to neighborhood ratio 

# display(under_25_data)

grouped = capac_data.groupby('geographic_cluster_name').agg(
    {'capacity': 'sum', 'total_population': 'mean'})

grouped['Total Capacity/Total Population Ratio'] = (grouped['capacity'] / grouped['total_population'])

grouped = grouped.drop(columns = ['capacity', 'total_population'])

grouped = grouped.sort_values('Total Capacity/Total Population Ratio', ascending=False)

display(grouped)

barplot_data = grouped[:5].append(grouped.tail(5))

# display(barplot_data)

# plot the DataFrame as a barplot
ax = barplot_data.plot(kind='bar', color = '#4E2A84')

# set the title and labels of the plot
ax.set_title('Capacity/Population Ratio for Different Chicago Neighborhoods')
ax.set_xlabel('Chicago Neighborhoods')
ax.set_ylabel('Total Capacity/Total Population Ratio')

# remove the legend from the plot
plt.legend([''], frameon=False)
              
# show the plot
plt.show()
```

```python
## pie chart of program category distribution 

# define a list of categories to keep
categ_to_keep = ['Reading & Writing', 'Science', 'Academic Support',
                'Computers', 'Social Studies', 'Managing Money']

# filter the DataFrame to keep rows with categories to keep
educ_prog_data = under_25_data[under_25_data['category_name'].isin(categ_to_keep)]

# count the frequency of each category
counts = educ_prog_data['category_name'].value_counts()

# create a pie chart
counts.plot.pie(autopct='%1.1f%%')

# set the title of the plot
plt.title('Educational Program Category Distribution') # under 25

# show the plot
plt.show()
```

```python
# Average length of programs based on geographic cluster (boxplot)

# Calculate the 1.5*IQR range for each group
q1 = under_25_data['program_length'].describe()['25%']
q3 = under_25_data['program_length'].describe()['75%']
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr 
upper_bound = q3 + 1.5 * iqr

# filter out outliers
length_data = under_25_data.loc[(under_25_data['program_length'] >= lower_bound) 
            & (under_25_data['program_length'] <= upper_bound)]

length_data = length_data[length_data['program_length'] > 0]

# Create a boxplot with grouped values
boxplot = length_data.boxplot(by='cluster_location', column='program_length', whis=75)

# Set plot title and labels
# Remove the plot title
plt.title('Boxplot of Program Lengths for Different Cluster Locations')
plt.xlabel('Cluster Location')
plt.ylabel('Program Length (days)')

# Rotate the x tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()

temp = length_data[length_data['cluster_location'] == 'Far Southeast Side']

temp['program_length'].median()
```
